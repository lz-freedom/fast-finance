```python
# 国家地区列表
regions = ["us", "cn", "sg", "hk", "in"]
# 盘中市值最小值
minIntradayMarketCap = 20000000
# 日交易量最小值
minDayVolume = 5000000
# 交易所
exchanges = ['SHH','SHZ','HKG','NYQ','NMS','SES','JPX','NSI','LSE','TOR','VAN','ASX']
# 不同国家地区返回股票条数
size = 25

data = {}
for region in regions:
    query = EquityQuery("and", [
        EquityQuery("eq", ["region", region]),
        EquityQuery("gte", ["intradaymarketcap", minIntradayMarketCap]),
        EquityQuery("gt", ["dayvolume", minDayVolume]),
        EquityQuery("is-in", ['exchange', *exchanges])
    ])
    resp = yf.screen(
        query,
        size=size,
        sortField="dayvolume",
        sortAsc=False
    )
    print(region)

    list = []
    for q in resp["quotes"]:
        print(q.get("symbol"), "|", q.get("displayName"), "|",q.get("shortName"),"|", q.get("prevName"))
        list.append(q)

    data[region] = list
print(data["us"][0])
```
